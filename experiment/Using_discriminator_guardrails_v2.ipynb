```python
import os
from getpass import getpass
from google.colab import userdata


os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
```

```python
yaml_content = """
models:
- type: main
  engine: openai
  model: gpt-3.5-turbo-0125
"""
colang_content = """

define user ask entertainment
    "suggest some good movies to watch"
    "Is Tom Hanks overrated?"
    "What is the plot of Swades movie?"
    "Why did people like about Stranger Things? "

define bot answer entertainment
    "I'm an education assistant. Will be happy to assist with your studies."


define flow entertainment
    user ask entertainment
    bot answer entertainment
"""
```

```python
from nemoguardrails import LLMRails, RailsConfig

config = RailsConfig.from_content(
    yaml_content = yaml_content,
    colang_content = colang_content
)

rails = LLMRails(config)
```

```python
res = await rails.generate_async(prompt = "Hey there!")
print(res)
```

```python
res = await rails.generate_async(prompt = "Which movies can I watch this weekend?")
print(res)
```
